3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def norma(matrix):
    matrix_sum = matrix.sum()
    return matrix / matrix_sum 

a = np.array([[1,2],[3,4]])
a = norma(a)
print(a)
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def create_matrix(n):
    matrix = np.zeros((n,n))
    matrix[0, :] = 1
    matrix[-1, :] = 1
    matrix[:, -1] = 1
    matrix[:, 0] = 1
    return matrix

arr = create_matrix(4)
print(arr)
    
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import matplotlib.pyplot as plt
merged_data = pd.merge(data, covid_data, on='location')

continent_cases = merged_data.groupby('continent')['total_cases'].sum().reset_index()

plt.figure(figsize=(5, 5))
plt.pie(
    continent_cases['total_cases'],
    labels=continent_cases['continent'],
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05]*len(continent_cases),
    shadow=True,
    colors=plt.cm.tab20.colors
)

plt.title('COVID-19 по континентам', fontsize=16)
plt.tight_layout()
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
import seaborn as sns

countries = ['Russia', 'China', 'United States', 'India', 'Nigeria']
tests_data = merged_data[merged_data['location'].isin(countries)][['location', 'total_tests']]
tests_data = tests_data.sort_values('total_tests', ascending=False)

barplot = sns.barplot(
    x='location',
    y='total_tests',
    data=tests_data,
    order=tests_data['location']  
)

plt.title('Число тестов на COVID-19')
plt.xlabel('Страна')
plt.ylabel('Общее число тестов')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.tight_layout()
plt.show()
```

3. Визуализируйте зависимость ВВП страны и случаев заболевания COVID-19

```python
plt.figure(figsize=(8, 6))

scatter = sns.regplot(
    x='gdp_per_capita',
    y='total_cases',
    data=merged_data,
    scatter_kws={'alpha':0.6, 'color':'blue'},
    line_kws={'color':'red', 'linestyle':'--'}
)

plt.title('Зависимость между ВВП и случаями COVID-19', fontsize=16, pad=20)
plt.xlabel('ВВП на душу населения (USD)', fontsize=12)
plt.ylabel('Случаев заболевания', fontsize=12)

plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${int(x):,}"))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))

plt.tight_layout()
plt.show()
```

4. Постройте гистограмму распределения количества смертей на один миллион человек

```python
merged_data = pd.merge(data, covid_data, on='location')
merged_data['deaths_by_million'] = 1000000 * merged_data['total_deaths'] / merged_data['population']

plt.figure(figsize=(14, 8))

# Строим гистограмму
sns.histplot(
    data=merged_data,
    x='deaths_by_million',
    bins=30,  
)

# Настройки оформления
plt.title('Распределение количества смертей от COVID-19 на 1 млн населения', fontsize=16)
plt.xlabel('Смертей на 1 млн населения', fontsize=12)
plt.ylabel('Количество стран', fontsize=12)

plt.tight_layout()
plt.show()
```
